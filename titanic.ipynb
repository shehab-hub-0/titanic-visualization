```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def load_data():
    data=pd.read_csv(r"C:\Users\a\Desktop\data_eng_project\assignment\Visualization\titanic\Titanic-Dataset.csv")
    data=data.drop(columns=["Cabin"])
    data["Age"]=data["Age"].fillna(data["Age"].mean())
    data["Embarked"]=data["Embarked"].fillna(data["Embarked"].mode()[0])
    data["Sex_numerical"]=data["Sex"].map({"male":0,"female":1})
    return data
```

```python
data=load_data()
```

```python
data
```

```python
sns.countplot(x='Survived',data=data)
plt.title("Survived VS NonSurvived")
plt.xlabel("0<--->Survived  | NonSurvived<--->1")
```

```python
sns.countplot(data=data, x='Sex', hue='Survived')
plt.title('Survived VS Sex')
plt.show()
```

```python
sns.countplot(data=data, x='Pclass', hue='Survived')
plt.title('Survival by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])

```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='Fare', hue='Survived', multiple='stack', bins=40)
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Count')
```
